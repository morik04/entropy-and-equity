CSV Generator

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
def create_csv(ticker_name, start_date, end_date):
    data = yf.download(ticker_name, start=start_date, end=end_date, auto_adjust=True)

    if not data.empty:
        # Flatten multi-index header that yfinance sometimes returns
        prices = data[["Close"]].rename(columns={"Close": "Adj Close"})
        prices.index.name = "Date"
        prices.to_csv(f"{ticker_name}.csv")
        print(f"Success! {ticker_name}.csv has been created.")
    else:
        print("No data found. Check ticker symbol.")
```

```python
def csv_reader(file_name):
    """
    Reads CSV, handles formatting, removes bad data, and returns aligned arrays.
    """
    df = pd.read_csv(file_name)
    
    # convert to date time objects
    if pd.api.types.is_numeric_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'], unit='D')
    else:
        df['Date'] = pd.to_datetime(df['Date'])

    # sorting by date
    df = df.sort_values('Date')

    df = df.dropna(subset=['Date', 'Adj Close'])

    # dates to int
    start_date = df['Date'].min()
    df['Days'] = (df['Date'] - start_date).dt.days

    x_data = df['Days'].values
    y_data = df['Adj Close'].values
    
    # return dates for plotting
    original_dates = df['Date'].values
    
    return x_data, y_data, original_dates
```
